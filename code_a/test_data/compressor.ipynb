```python
import os
import shutil
import zipfile

# 定义文件夹路径
source_folder = 'c_pred'
target_folder = os.path.join(source_folder, 'track_c')
zip_file_path = os.path.join(source_folder, 'track_c.zip')

# 创建 target_folder 文件夹，如果不存在的话
if not os.path.exists(target_folder):
    os.makedirs(target_folder)

# 复制所有 CSV 文件到 target_folder
for file_name in os.listdir(source_folder):
    if file_name.endswith('.csv'):
        source_file = os.path.join(source_folder, file_name)
        target_file = os.path.join(target_folder, file_name)
        shutil.copy(source_file, target_file)
        print(f"Copied {file_name} to {target_folder}")

# 压缩 target_folder 为 track_a.zip
with zipfile.ZipFile(zip_file_path, 'w', zipfile.ZIP_DEFLATED) as zipf:
    for root, dirs, files in os.walk(target_folder):
        for file in files:
            zipf.write(os.path.join(root, file), os.path.relpath(os.path.join(root, file), target_folder))
            print(f"Added {file} to zip")

print(f"Track folder {target_folder} has been compressed into {zip_file_path}")
```
