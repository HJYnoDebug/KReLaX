```python
import pandas as pd

# 读取CSV文件
df = pd.read_csv("mergedA.csv")

# 确定不需要计算的列
exclude_columns = ['id', 'text', 'lang']

# 获取所有标签列（假设标签列是以0和1为值的，其它列除了排除的列）
label_columns = [col for col in df.columns if col not in exclude_columns]

# 按 'lang' 列进行分组
grouped = df.groupby('lang')

# 计算每个子集中的标签0与1的比例
result = {}

for lang, group in grouped:
    lang_result = {}
    for label in label_columns:
        label_counts = group[label].value_counts(normalize=True)  # 计算标签的频率（比例）
        lang_result[label] = label_counts.to_dict()  # 将结果转化为字典形式
    result[lang] = lang_result

# 输出结果
for lang, lang_result in result.items():
    print(f"Lang: {lang}")
    for label, counts in lang_result.items():
        print(f"  Label '{label}':")
        for value, proportion in counts.items():
            print(f"    Value {value}: {proportion:.2f}")
    print("\n")
```

```python
import os
import pandas as pd
import random

# 读取 mergedA.csv 文件
df = pd.read_csv("mergedA.csv")

# 确定需要排除的列
exclude_columns = ['id', 'text', 'lang']

# 获取标签列（排除id, text, lang列）
label_columns = [col for col in df.columns if col not in exclude_columns]

# 将所有列名转换为小写，确保大小写不敏感
df.columns = df.columns.str.lower()

# 计算标签的 0 和 1 比例
def compute_label_proportions(group):
    proportions = {}
    for label in label_columns:
        label_counts = group[label].value_counts(normalize=True)  # 计算标签的比例
        proportions[label] = label_counts.to_dict()
    return proportions

# 获取 lang 列的所有唯一值（每个 objective 子集）
langs = df['lang'].unique()

# 新增增强数据集
augmented_df = pd.DataFrame()

# 设置最大循环次数
max_iterations = 5

# 循环直到满足条件
for lang in langs:
    # 获取 lang 对应的子集
    lang_group = df[df['lang'] == lang]
    
    # 计算每个标签的 0 和 1 比例
    proportions = compute_label_proportions(lang_group)
    
    # 计算哪些标签 1 的比例小于 10%
    labels_to_augment = [label for label, prop in proportions.items() if prop.get(1, 0) < 0.1]

    iteration_count = 0
    while labels_to_augment and iteration_count < max_iterations:
        iteration_count += 1
        
        # 找到与 lang 相同的 CSV 文件
        augment_file_path = f"../aug_data/{lang}.csv"
        
        if os.path.exists(augment_file_path):
            augment_df = pd.read_csv(augment_file_path)
            
            # 将增强数据文件的所有列名转换为小写
            augment_df.columns = augment_df.columns.str.lower()
            
            # 只选择标签为 1 的样本
            for label in labels_to_augment:
                if label in augment_df.columns:
                    augment_label_df = augment_df[augment_df[label] == 1]
                    
                    # 如果标签为 1 的样本数量不足
                    num_samples_needed = int(len(lang_group) * 0.1) - lang_group[label].sum()
                    if num_samples_needed > 0:
                        if len(augment_label_df) < num_samples_needed:
                            # 如果增强数据中的标签为1的样本不足，抽取全部样本
                            augment_samples = augment_label_df
                        else:
                            # 如果有足够样本，随机抽取所需数量的样本
                            augment_samples = augment_label_df.sample(n=num_samples_needed, random_state=42)
                        
                        # 将 lang 列添加到增强数据集
                        augment_samples['lang'] = lang  # 设置当前样本的 lang 列为当前语言

                        # 将这些样本添加到增强数据集中
                        augmented_df = pd.concat([augmented_df, augment_samples], ignore_index=True)

                        # 添加完后去重
                        augmented_df = augmented_df.drop_duplicates(subset=['text'])
                        
                        # 重新计算标签的 0 和 1 比例
                        proportions = compute_label_proportions(augmented_df)
                        
                        # 重新检查哪些标签需要增强
                        labels_to_augment = [label for label, prop in proportions.items() if prop.get(1, 0) < 0.1]
                        
                        # 如果已经满足所有标签的比例，跳出循环
                        if not labels_to_augment:
                            break
        else:
            print(f"Warning: {augment_file_path} does not exist!")
            break  # 如果找不到对应的文件，就跳出循环

    # 如果五次循环仍然没有达到目标，结束循环
    if iteration_count == max_iterations and labels_to_augment:
        print(f"Warning: Unable to achieve 10% for labels in lang: {lang} after {max_iterations} iterations.")

# 如果有增强数据，保存到新的文件
if not augmented_df.empty:
    augmented_df.to_csv("aug_mergedA.csv", index=False)
    print("New augmented dataset saved as 'aug_mergedA.csv'.")
else:
    print("No new augmented data added.")
```

```python
import pandas as pd

# 读取aug_mergedA文件
aug_mergedA_path = 'aug_mergedA.csv'
data = pd.read_csv(aug_mergedA_path)

# 检查'id'列是否存在空缺值
if data['id'].isnull().any():
    print("There are missing values in the 'id' column. Filling them now.")

    # 获取所有 lang 列的唯一值
    lang_values = data['lang'].unique()
    
    # 生成新的 id 列：以 _train_track_a_00001 这种格式生成
    new_ids = []
    for lang in lang_values:
        lang_subset = data[data['lang'] == lang]
        lang_id_range = len(lang_subset)
        for i in range(lang_id_range):
            new_ids.append(f"{lang}_train_track_a_{str(i+1).zfill(5)}")
    
    # 将生成的 id 填充到 data 中
    data['id'] = new_ids

# 调整列顺序
desired_order = ['id', 'text'] + [col for col in data.columns if col not in ['id', 'text', 'lang']] + ['lang']
data = data[desired_order]

# 保存更新后的文件
data.to_csv(aug_mergedA_path, index=False)

print("Missing 'id' values have been filled, and the file has been updated.")
```
